```python
from ipywidgets import StaticInteract, RangeWidget, RadioWidget

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#var = 1

def plot(var):
    fig, ax = plt.subplots(figsize=(4, 3),subplot_kw={'axisbg':'#EEEEEE','axisbelow':True})
    ax.grid(color='w', linewidth=2, linestyle='solid')
    #ax.legend(loc='upper right')
    x = np.linspace(-2, 2, 201)
    y = (1./(var*(2*np.pi)**(1./2)))*np.exp(-x**2/(2*var**2))
    ax.plot(x, y, lw=5, alpha=0.4,label=var)
    return fig

StaticInteract(plot,var=RangeWidget(0.1, 1, 0.1))
```

```python
from ipywidgets import StaticInteract, RangeWidget, RadioWidget

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#var = 1

def plot(var):
    fig, ax = plt.subplots(figsize=(4, 3),subplot_kw={'axisbg':'#EEEEEE','axisbelow':True})
    ax.grid(color='w', linewidth=2, linestyle='solid')
    #ax.legend(loc='upper right')
    x = np.linspace(-2, 2, 201)
    y = np.fft.fft((1./(var*(2*np.pi)**(1./2)))*np.exp(-x**2/(2*var**2)))
    ax.plot(y, lw=5, alpha=0.4,label=var)
    #ax.set_xlim(150,200)
    return fig

StaticInteract(plot,var=RangeWidget(0.01, 0.1, 0.01))
```

```python
st = np.zeros(101) # saw tooth function
for i in xrange(len(st)):
    st[i] = i%9
```

```python
tr = np.zeros(101) # triangular function
for i in xrange(len(tr)):
    if i%20 < 10:
        tr[i] = i%10
    if i%20 > 9:
        tr[i] = 10-i%10
```

```python
plt.subplot(121)
plt.plot(np.fft.fft(st))
plt.subplot(122)
plt.plot(np.fft.fft(tr))
```
